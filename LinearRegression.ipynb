# LINEAR REGRESSION

```python
#import required librarie

%matplotlib inline
import numpy as np
from pylab import *
```

```python
#Fabricate random data that shows linear relationship between page_speeds and amount_purchased
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0.0, 0.1, 1000)) * 3
scatter(pageSpeeds, purchaseAmount)
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_error = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
#output the r_value score
r_value ** 2
```

```python
#use intercept and slope values to plot a line of predicted values vs observed
import matplotlib.pyplot as plt
def predict(x):
    return slope * x + intercept
fitLine = predict(pageSpeeds)
plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c = 'r')
```
